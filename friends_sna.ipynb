```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()

with open("friends_episodes.txt", "r") as f:
    for line in f:
        if line.startswith("#"):
            continue

        parts = line.strip().split()
        if len(parts) != 2:
            continue

        u, v = parts

        if G.has_edge(u, v):
            G[u][v]['weight'] += 1
        else:
            G.add_edge(u, v, weight=1)

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```
